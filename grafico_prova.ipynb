```python
import pandas as pd
import matplotlib.pyplot as plt

#Carica il file CSV
file_path = "Dataset/1- mental-illnesses-prevalence.csv"  # <-- sostituisci con il nome del nuovo file
df = pd.read_csv(file_path)

#Filtra gli anni dal 1990 al 2019
df = df[(df["Entity"] == "America (IHME GBD)")]
df = df[(df["Year"] >= 1990) & (df["Year"] <= 2019)]

#Definisci le colonne corrispondenti ai problemi mentali
colonne_disturbi = {
    "Schizofrenia": "Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Depressione": "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Ansia": "Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Bipolarit": "Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized",
    "Problemi d'alimentazione": "Eating disorders (share of population) - Sex: Both - Age: Age-standardized",
}

#Calcola la media globale per ogni anno
df_media = df.groupby("Year")[[col for col in colonne_disturbi.values()]].mean().reset_index()

#Crea il grafico
plt.figure(figsize=(12, 6))
for nome, col in colonne_disturbi.items():
    plt.plot(df_media["Year"], df_media[col], label=nome)

plt.title("Variazione percentuale di popolazione affetta da disturbi mentali (1990-2019)")
plt.xlabel("Anno")
plt.ylabel("Percentuale (media globale)")
plt.legend()
plt.grid(True)
plt.show()

 
```
