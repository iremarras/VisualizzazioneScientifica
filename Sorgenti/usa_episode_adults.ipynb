```python
import pandas as pd
import matplotlib.pyplot as plt

# Dati
data = {
    "Category": ["Overall", "Female", "Male", "18-25", "26-49", "50+"],
                 
    "Group": ["Overall", "Sex", "Sex", "Age", "Age", "Age"],
    "Percent": [8.3, 10.3, 6.2, 18.6, 9.3, 4.5]
}

df = pd.DataFrame(data)

# Palette colori 
colors = {
    "Overall": "#2e7d32",
    "Sex": "#c8e6c9",
    "Age": "#66bb6a",
    "Race/Ethnicity": "#e8f5e9"
}


# Ordinare il dataframe
group_order = ["Overall", "Sex", "Age"]
df["Group"] = pd.Categorical(df["Group"], categories=group_order, ordered=True)
df = df.sort_values("Group")


# Creazione del grafico
plt.figure(figsize=(8,6))
bars = plt.bar(df["Category"], df["Percent"], color=[colors[g] for g in df["Group"]], edgecolor='none')

bar_width = 0.3   
bar_gap = 0.05      

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f"{height:.1f}", 
             ha="center", va="bottom", fontsize=10, color="#333333")

# Titolo e assi
plt.title("Past Year Prevalence of Major Depressive Episodes\nAmong U.S. Adults (2021)", fontsize=14)

plt.ylabel("Percent (%)", fontsize=12)
plt.ylim(0, 35)

# Legenda
handles = [plt.Rectangle((0,0),1,1, color=colors[group]) for group in group_order]
plt.legend(handles, group_order, 
           title="Group", 
           frameon=False, 
           bbox_to_anchor=(1.05, 1),  
           loc='upper left')          

plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout(pad=2)


plt.show()

```
