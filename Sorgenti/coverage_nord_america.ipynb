```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file
file_path = "Dataset/coverage.csv"  
df = pd.read_csv(file_path)

# Parametri da modificare
nazione = "North America"


# Filtra i dati per nazione e anno
df_sel = df[(df["Entity"] == nazione)].iloc[0]

# Colonne delle prevalenze
colonne_disturbi = {
    "Depessive disorders": "Major depression",
    "Schizophrenia": "Schizophrenia",
    "Bipolar disorder": "Bipolar disorder",
    "Eating disorders": "Eating disorders",
    "Anxiety disorders": "Anxiety disorders",
}

# Estraggo i valori 
valori = {nome: float(df_sel[col]) for nome, col in colonne_disturbi.items()}
serie = pd.Series(valori)

# Colori personalizzati
colori_pastello = ["#80c680", "#ffb26e", "#bfa4d7", "#79add2", "#e67d7e"] 

# Grafico a barre orizzontali
plt.figure(figsize=(5,3))
serie.plot(kind="barh", color=colori_pastello, edgecolor="black", width=0.5, zorder=2)

plt.title(f"Prevalence of mental disorders in {nazione}")
plt.xlabel("Population (%))")
plt.ylabel("Mental disorder")
plt.grid(axis="x", linestyle="-", alpha=0.7, zorder=0)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)

plt.show()
```
