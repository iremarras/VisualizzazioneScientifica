```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file
df = pd.read_csv("Dataset\eu_deaths.csv")

# ordino per anno
df = df.sort_values("TIME_PERIOD")

# Plot
plt.figure(figsize=(10,5))
plt.plot(df["TIME_PERIOD"], df["OBS_VALUE"], marker="o", color="steelblue", linewidth=2)

# etichette e titolo
plt.title("Deaths from self-harm (2011–2022)", fontsize=18)
plt.xlabel("Year", fontsize= 16)
plt.ylabel("Death rate (per 100.000 inhabitants)", fontsize= 16)
plt.grid(True, linestyle="--", alpha=0.6)
plt.ylim(top=15)
plt.ylim(bottom=0)

plt.tight_layout()
plt.show()

# Variazione
valore_iniziale = df.loc[df["TIME_PERIOD"] == 2013, "OBS_VALUE"].values[0]
valore_finale   = df.loc[df["TIME_PERIOD"] == 2022, "OBS_VALUE"].values[0]

differenza = valore_finale - valore_iniziale
var_percentuale = (differenza / valore_iniziale) * 100

print(f"Nel periodo 2013–2022 il valore è cambiato di {differenza:.2f} ({var_percentuale:.1f}%).")
```
