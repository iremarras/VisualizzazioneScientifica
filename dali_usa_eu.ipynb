```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica i dati
file_path = "Dataset/daly_eu_us.csv"
df = pd.read_csv(file_path)

# Filtra solo Western Europe e USA, entrambi i sessi e tutte le età
df_sel = df[(df["location"].isin(["Western Europe", "United States of America"])) &
            (df["sex"] == "Both") & (df["age"] == "All ages")]

# Ordina per anno
df_sel = df_sel.sort_values(["location", "year"])

colors = ["blue", "orange"]

# Plot
plt.figure(figsize=(8,5))

for i, (loc, data) in enumerate(df_sel.groupby("location")):
    plt.plot(data["year"], data["val"], label=loc, color=colors[i])

plt.title("DALYs from Depressive Disorders (Rate per 100,000 inhabitants)")
plt.xlabel("Year")
plt.ylabel("DALYs rate")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(title="Location", fontsize=10)
plt.tight_layout()
plt.show()
```
